```python
import pandas as pd
```

```python
db_target = pd.read_csv('EUR_USD Historical Data 3 anni al 28 ottobre 2019.csv')
```

```python
db_target
```

```python
db_target['Change %'] = db_target['Change %'].str.replace('%', '')
db_target['Change %'] = db_target['Change %'].astype(float)
```

```python
db_target['Target'] = (db_target['Change %'] > 0).astype(int)
```
